```python
import pandas as pd
import csv
import os
import re
from pathlib import Path
```

```python
def read_folders(femur_path, tibia_path, sipla_path):
    femurs = sorted([int(re.match(r'kneefit_femur_(\d+)_syn.png', i).group(1)) for i in os.listdir(femur_path)])
    tibias = sorted([int(re.match(r'kneefit_tibia_(\d+)_syn.png', i).group(1)) for i in os.listdir(tibia_path)])

    sipla = pd.read_csv(sipla_path)
    siplas = sipla['id'].astype(int).tolist()

    print(siplas)

    print(f"{femur_path} has {len(femurs)} images")
    print(f"{tibia_path} has {len(tibias)} images")

    matching_anatomies = list(set(siplas) & set(femurs) & set(tibias))
    print(f"there are {len(matching_anatomies)} matches")

    return matching_anatomies


def generate_csv(matches, filename, femur_path, tibia_path, sipla_path):
    femur_list = [os.path.join(Path(femur_path), f"kneefit_femur_{id}_syn.png") for id in matches]
    tibia_list = [os.path.join(Path(tibia_path), f"kneefit_tibia_{id}_syn.png") for id in matches]

    print(femur_list)
    print(tibia_list)

    sipla = pd.read_csv(sipla_path)
        # Ensure that the 'id' column is of string type to match the match_list
    sipla['id'] = sipla['id'].astype(int)

    # Filter the DataFrame to get rows where 'id' is in match_list
    sipla_list = sipla[sipla['id'].isin(matches)]["img_path"].tolist()

    df = pd.DataFrame({
        "xray": sipla_list,
        "femur": femur_list,
        "tibia": tibia_list
    })

    df.to_csv(filename, header=False, index=False)


```

```python
femur_path=r"D:\\kneefit_femur_synthetic"
tibia_path=r"D:\\kneefit_tibia_synthetic"
sipla_path = r"C:\\Users\\Public Admin\\Desktop\\Gitlab\\kneedeeppose\\data\\sipla_bone_local_preprocessed.csv"
csv_path = r"C:\\Users\\Public Admin\\Desktop\\Gitlab\\kneedeeppose\\data\\segnet_data.csv"


matches = read_folders(femur_path, tibia_path, sipla_path)

generate_csv(matches, csv_path, femur_path, tibia_path, sipla_path)
```
