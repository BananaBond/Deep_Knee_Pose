## Replace paths

```python
paths_to_replace = {
    "img_path": "D:\kneefit_images_full_resolution",
    "femur_stl": "D:\kneefit_model",
    "femur_nii": "D:\kneefit_model",
    "tibia_stl": "D:\kneefit_model",
    "tibia_nii": "D:\kneefit_model"
}
```

check if the files are there in the replacement path

```python
uniques = {}
for path in paths_to_replace.keys():
    uniques[path] = []
    uniques[path].extend(knee_df[path].unique())    
    # print(Path(paths_to_replace[path]))
    # print(Path("vv1/vvv/vv2").parts[-1])

    uniques[path] = [os.path.join(paths_to_replace[path], Path(i).parts[-1].replace("tiff", "tif")) for i in uniques[path]]
    path_exists = [i for i in uniques[path] if not os.path.exists(i)]

    print(f"{len(path_exists)} missing files in {path} replacement path")
```

```python
uniques
```

```python
for path in paths_to_replace.keys():
    grouped = knee_df.groupby(path).groups

    for i, (key, indices) in enumerate(grouped.items()):
        # print(f"Group '{key}': {indices}")
        knee_df.loc[indices, path] = uniques[path][i]
```

## More preprocessing, to add the spacing in nii files

```python
import nibabel as nib

def get_spacing(path):
    nii_data = nib.load(path)
    # Access header information
    header = nii_data.header
    return list(header.get_zooms())
```

```python
knee_df = pd.read_csv(r"C:\Users\Public Admin\Desktop\Gitlab\kneedeeppose\data\sipla_bone_local_preprocessed.csv")

patients = knee_df["patient_id"].unique()
print(patients)

for i in patients:
    df = knee_df[knee_df["patient_id"] == i]
    print(df.shape[0])
    volume_path = df["femur_nii"].unique()
    assert len(volume_path) == 1
    spacing = get_spacing(volume_path[0])
    print(spacing)

    df['spacing'] = [spacing for _ in range(df.shape[0])]  # Ensure the length of the new column matches the filtered subset

    # Convert lists to strings
    df['spacing'] = df['spacing'].apply(lambda x: str(x))

    print(df['spacing'])

    # Update the original DataFrame with the new column values
    knee_df.loc[knee_df["patient_id"] == i, 'spacing'] = df['spacing'].values

```
